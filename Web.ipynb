```python
import requests
from bs4 import BeautifulSoup
import threading
import multiprocessing
import asyncio
import time

def scrap(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    title = soup.title.text
    paragraphs = [paragraph.text for paragraph in soup.find_all('p')]
    links = [link.get('href') for link in soup.find_all('a')]
    return {'title': title, 'paragraphs': paragraphs, 'links': links}

def threaded_worker(url, results, lock):
    data = scrap(url)
    with lock:
        results[url] = data

def scrape_parallel_threading(urls, num_threads):
    results = {}
    lock = threading.Lock()

    def worker(url):
        threaded_worker(url, results, lock)

    threads = []
    for url in urls:
        thread = threading.Thread(target=worker, args=(url,))
        thread.start()
        threads.append(thread)

    for thread in threads:
        thread.join()

    return results

def multiprocessing_worker(args):
    url, results, lock = args
    data = scrap(url)
    with lock:
        results[url] = data

def scrape_parallel_multiprocessing(urls, num_processes):
    results = {}
    manager = multiprocessing.Manager()
    lock = manager.Lock()

    pool_args = [(url, results, lock) for url in urls]

    with multiprocessing.Pool(processes=num_processes) as pool:
        pool.map(multiprocessing_worker, pool_args)

    return results

async def asyncio_worker(url, results, lock):
    data = scrap(url)
    async with lock:
        results[url] = data

async def scrape_parallel_asyncio(urls):
    results = {}
    manager = multiprocessing.Manager()
    lock = manager.Lock()

    tasks = [asyncio_worker(url, results, lock) for url in urls]

    await asyncio.gather(*tasks)

    return results

def measure_execution_time(func, *args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    execution_time = end_time - start_time
    return result, execution_time

# Example usage:
urls_to_scrape = ["https://example.com", "https://www.saucedemo.com/", "https://ultimateqa.com"]

num_cores_options = [1, 2, 4, 8]

for num_cores in num_cores_options:
    print(f"\nNumber of Cores: {num_cores}")

    print("\nUsing Threading:")
    try:
        results_threading, execution_time_threading = measure_execution_time(scrape_parallel_threading, urls_to_scrape, num_cores)
        print(f"Execution Time: {execution_time_threading} seconds")
    except Exception as e:
        print(f"An exception occurred during threading: {e}")

    print("\nUsing Multiprocessing:")
    try:
        results_multiprocessing, execution_time_multiprocessing = measure_execution_time(scrape_parallel_multiprocessing, urls_to_scrape, num_cores)
        print(f"Execution Time: {execution_time_multiprocessing} seconds")
    except Exception as e:
        print(f"An exception occurred during multiprocessing: {e}")

    print("\nUsing Asyncio:")
    try:
        results_asyncio, execution_time_asyncio = measure_execution_time(scrape_parallel_asyncio, urls_to_scrape)
        print(f"Execution Time: {execution_time_asyncio} seconds")
    except Exception as e:
        print(f"An exception occurred during asyncio: {e}")
```

```python
%%writefile filename.C
#include <stdio.h>
#include <string.h>
#include <stdlib.h>
#include <omp.h>

#define MAX_PARAGRAPHS 10
#define MAX_PARAGRAPH_LENGTH 200

// Structure to hold scraped data
struct ScrapedData {
    char paragraphs[MAX_PARAGRAPHS][MAX_PARAGRAPH_LENGTH];
    int num_paragraphs;
};

// Function to scrape data from the web (simulated)
void scrape_web(struct ScrapedData *data) {
    // Simulate scraping paragraphs from a website
    data->num_paragraphs = 5;
    for (int i = 0; i < data->num_paragraphs; ++i) {
        snprintf(data->paragraphs[i], MAX_PARAGRAPH_LENGTH, "Paragraph %d", i + 1);
    }
}

int main() {
    struct ScrapedData result;
    struct ScrapedData total_result;

    // Initialize total_result with zeros
    memset(&total_result, 0, sizeof(total_result));

    // Simulate scraping paragraphs from the web in parallel
    #pragma omp parallel for private(result) shared(total_result)
    for (int i = 0; i < 100; ++i) {
        // Each thread has its own result
        scrape_web(&result);

        // Use a critical section to update total_result
        #pragma omp critical
        {
            // Copy the paragraphs to total_result
            for (int j = 0; j < result.num_paragraphs; ++j) {
                strncpy(total_result.paragraphs[total_result.num_paragraphs],
                        result.paragraphs[j], MAX_PARAGRAPH_LENGTH);
                total_result.num_paragraphs++;
            }
        }
    }

    // Print or process the total_result
    printf("Total Number of Paragraphs: %d\n", total_result.num_paragraphs);
    for (int i = 0; i < total_result.num_paragraphs; ++i) {
        printf("Paragraph %d: %s\n", i + 1, total_result.paragraphs[i]);
    }

    return 0;
}

```
