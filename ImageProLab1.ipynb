```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


img = Image.open('yoshi.png')
gray_img = img.convert('L')
norm_img = np.array(gray_img) / 255.0
c = 40.95
log_img = np.log(1 + c * norm_img)
log_img = (log_img - np.min(log_img)) / (np.max(log_img) - np.min(log_img))
plt.imshow(log_img, cmap='gray')
plt.axis('off')
plt.title('Log-transformation ')
plt.show()

```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


img = Image.open('yoshi.png')
gray_img = img.convert('L')
norm_img = np.array(gray_img) / 255.0

gamma = 0.8
power_img = np.power(norm_img, gamma)

power_img = (power_img - np.min(power_img)) / (np.max(power_img) - np.min(power_img))

plt.imshow(power_img, cmap='gray')
plt.axis('off')
plt.title('Power-transformation ')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Load the color image
color_img = cv2.imread('yoshi.png')

# Convert the color image to grayscale
gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)

# Perform the threshold transformation
thresh_value = 90
ret, thresh_img = cv2.threshold(gray_img, thresh_value, 255, cv2.THRESH_BINARY)



# Display the images
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
plt.title('Color Image')

plt.subplot(1, 3, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(1, 3, 3)
plt.imshow(thresh_img, cmap='cool')
plt.title('Threshold Image')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Load the image
img = cv2.imread('yoshi.png')

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Determine the minimum and maximum pixel values
M1 = 100
M2 = 170

stretch_img = ((gray_img - M1) / (M2 - M1)) * 255
stretch_img = stretch_img.clip(0, 255).astype('uint8')


# Display the images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(stretch_img, cmap='PuOr')
plt.title('Contrast Stretched Image')

plt.show()

```
