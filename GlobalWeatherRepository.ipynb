This code will create a horizontal bar plot for the ten cities with the lowest temperatures on the left and the ten cities with the highest temperatures on the right.

```python
lowest_temp_df = df.sort_values(by = 'temperature_celsius', ascending=True).head(10)
highest_temp_df = df.sort_values(by='temperature_celsius', ascending=False).head(10)

fig , (ax1,ax2) = plt.subplots(1,2, figsize=(14,6))

ax1.barh(lowest_temp_df['location_name'], lowest_temp_df['temperature_celsius'], color='skyblue')
ax1.set_xlabel('Temperature (°C)')
ax1.set_title('Ten Cities with Lowest Temperatures (°C)')

ax2.barh(highest_temp_df['location_name'], highest_temp_df['temperature_celsius'], color='salmon')
ax2.set_xlabel('Temperature (°C)')
ax2.set_title('Ten Cities with Highest Temperatures (°C)')

plt.tight_layout()
plt.show()

```

This code should generate a line chart showing the daily average temperatures over time for each city in the Middle_East dataset.

```python
agg_sunrise_df = Middle_East.groupby(['location_name', 'sunrise'])[['temperature_celsius']].mean().reset_index()
agg_sunset_df = Middle_East.groupby(['location_name', 'sunset'])[['temperature_celsius']].mean().reset_index()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8))

for city in Middle_East['location_name'].unique():
    sunrise_data = agg_sunrise_df[agg_sunrise_df['location_name'] == city]
    sunset_data = agg_sunset_df[agg_sunset_df['location_name'] == city]

    ax1.plot(sunrise_data['sunrise'], sunrise_data['temperature_celsius'], label=city)
    ax2.plot(sunset_data['sunset'], sunset_data['temperature_celsius'], label=city)


ax1.set_title('Daily Average Temperatures at Sunrise (Middle East Cities)')
ax1.set_xlabel('Time at Sunrise')
ax1.set_ylabel('Temperature (°C)')
ax1.legend(loc='upper left')

ax2.set_title('Daily Average Temperatures at Sunset (Middle East Cities)')
ax2.set_xlabel('Time at Sunset')
ax2.set_ylabel('Temperature (°C)')
ax2.legend(loc='upper left')

ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

This code will generate a bar chart that compares the average humidity levels across different cities in the Middle_East dataset.

```python
agg_humidity_df = Middle_East.groupby(['location_name'])[['humidity']].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(agg_humidity_df['location_name'], agg_humidity_df['humidity'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Average Humidity (%)')
plt.title('Average Humidity Levels in Middle East Cities')
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```

```python
agg_co_df = Middle_East.groupby(['location_name'])[['air_quality_Carbon_Monoxide']].mean().reset_index()

cities = agg_co_df['location_name']
co_levels = agg_co_df['air_quality_Carbon_Monoxide']

plt.figure(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(cities))

plt.bar(index, co_levels, bar_width, label='CO Levels (Carbon Monoxide)', color='skyblue')

plt.xlabel('City')
plt.ylabel('CO Levels (Carbon Monoxide)')
plt.title('Comparison of Carbon Monoxide (CO) Levels in Middle East Cities')
plt.xticks(index, cities, rotation=45)
plt.tight_layout()

plt.legend()
plt.show()
```

```python
agg_air_quality_df = Middle_East.groupby(['location_name'])[['air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide']].mean().reset_index()

cities = agg_air_quality_df['location_name']
no2_levels = agg_air_quality_df['air_quality_Nitrogen_dioxide']
so2_levels = agg_air_quality_df['air_quality_Sulphur_dioxide']

plt.figure(figsize=(12, 6))
bar_width = 0.35
index = np.arange(len(cities))

plt.bar(index - bar_width/2, no2_levels, bar_width, label='NO2 Levels', color='skyblue', alpha=0.7)
plt.bar(index + bar_width/2, so2_levels, bar_width, label='SO2 Levels', color='orange', alpha=0.7)

plt.xlabel('City')
plt.ylabel('Air Quality Measurement')
plt.title('Comparison of NO2 and SO2 Levels in Middle East Cities')
plt.xticks(index, cities, rotation=45)
plt.tight_layout()


plt.legend()
plt.show()
```

air_quality_Nitrogen_dioxide: Air quality measurement: Nitrogen Dioxide

air_quality_Sulphur_dioxide: Air quality measurement: Sulphur Dioxide

```python
selected_features = ['temperature_celsius', 'humidity', 'air_quality_Carbon_Monoxide', 'air_quality_Ozone', 'air_quality_PM2.5', 'air_quality_PM10', 'air_quality_gb-defra-index', 'air_quality_us-epa-index']
cities = ['Doha', 'Abu Dhabi', 'Riyadh']
for city in cities:
  city_data = Middle_East[Middle_East['location_name'] == city]
  air_cor = city_data[selected_features]
  correlation_matrix = air_cor.corr()
  plt.figure(figsize=(15,10))
  sns.heatmap(correlation_matrix, annot=True, fmt=".1f",linewidths=0.5)
  plt.title(f'Heatmap for {city}')
  plt.xticks(rotation=80)
  plt.yticks(rotation=0)
  plt.tight_layout()
  plt.show()
```
